##Downloaded perceived need for mental health care dataset from Health Canada website (https://open.canada.ca/data/en/dataset/08e0af90-1737-4955-a6f0-c961acd2b466/resource/94e60f52-2f81-453b-8644-6aacea1431b0) and importing dataset here.

```python
import pandas as pd
df = pd.read_csv("Perceived_need.csv")
```

```python
df.head()
```

```python
len(df)
```

```python
df.describe(include='all')
```

###All data were of year 2018.

###Keeping only necessary variables

```python
df = df.drop(columns=['REF_DATE'])
df = df.drop(columns=['DGUID'])
df = df.drop(df.columns[df.columns.get_loc('UOM_ID'):df.columns.get_loc('COORDINATE')+1], axis=1)
df = df.drop(df.columns[df.columns.get_loc('STATUS'):df.columns.get_loc('DECIMALS')+1], axis=1)
```

```python
df.head()
```

###Limiting data to number of persons, and unmet perceived needs

```python
df = df.loc[(df['Characteristics'] == 'Number of persons')]
df = df.loc[(df['Indicators'] == 'Perceived need for mental health care, needs not met')]
```

```python
len(df)
```

```python
df.head()
```

###getting summary stattistics for unmet mental health needs by age groups, sex and geographical location

```python
df.groupby('Age group')['VALUE'].describe()
```

###18-34 years old had the highest unmet mental health care needs.

```python
df.groupby('Sex')['VALUE'].describe()
```

###Females had higher unmet mental health care needs than males.

####Plotting distribution of unmet mental health care needs by geographical locations

```python
import matplotlib.pyplot as plt
stats_by_geo = df[df['GEO'] != 'Canada (excluding territories)'].groupby('GEO')['VALUE'].describe()
plt.figure(figsize=(10, 6))
stats_by_geo['mean'].plot(kind='bar', color='skyblue', alpha=0.8) 
plt.title('Unmet mental health care needs across Canada')
plt.xlabel(' ')
plt.ylabel('Average number of persons with unmet needs')
plt.tight_layout()
plt.show()
```

###Ontario had the highest unmet mental health care needs.
